# A2 Text Adventure Game Development - The Hunger Games Part 1

This is a game called "Hunger Games" which takes place in Panem. This game simulates a brief summary of the brutal competition in which two tributes from each district (part of the country) have to fight to the death. In this script, the player takes on the role of either the male or female tribute from District 12 and must train and fight to survive. The target of this is to prepare the player for the Hunger Games and close with the final decision of the movie. The game includes several functions, including a print out of the other tributes, three stages to win the game, and outputs of wins or loses. Additionally implemented, there is some basic error handling to ensure that input is valid and that the game does not get stuck in an infinite loop. Mostly all patterns of the course are integrated and there is no indication of bugs.

```python
### imports
import random
import sys


# Functions
# Function used to create one space in output
def space():
    print()
    return

# Print the output for winners
def win():
    print(f"""
    =======================================
    |                                     |
    |                                     |
    |         Great Job, you won!         |
    |                                     |
    |                                     |
    =======================================""")
    return

# Print the output for loosers
def loose():
    print(f"""
    =======================================
    |                                     |
    |                                     |
    |      Nice try, but you lost...      |
    |                                     |
    |                                     |
    =======================================""")
    sys.exit()
    return

# Print all tributes out of a list. (Not all names are known,)
def print_opponents():
## List containing all tributes
    tributes = ["Glimmer", "Marvel", 
                "Clove", "Cato", 
                "Tribute Girl", "Tribute Boy",
                "Tribute Girl", "Tribute Boy",
                "Foxface","Tribute Boy",
                "Tribute Girl", "Tribute Boy",
                "Tribute Girl", "Tribute Boy",
                "Tribute Girl", "Tribute Boy",
                "Tribute Girl", "Tribute Boy",
                "Tribute Girl", "Tribute Boy",
                "Rue", "Thresh"]
    
    print("You have very strong opponents! Unfortunately I don’t know all of them yet, but here is everything I know.")

## for loop to iterate over the while list with a foratted output
    for x in range(11) :
        print(f"""District {x + 1 }: {tributes.pop(0)}, {tributes.pop(0)}""")

# Introduction Function to introduce the player and the game
def intro():
    print(f"""Welcome to the Hunger Games in Panem!

You’re in District 12 the poorest and most remote district of Panem. 
Each year, the Capitol, runs the cruel and exiting Hunger Games. 
Two out of each district fights to the death in a specially designed arena, 
until only one is left standing. 
You are one of the chosen ones of District 12 to fight for life! 
I'll help you to get prepared until you reach the arena""")

    space()
    
## input the username
    user_name = input(prompt = "What's your name? " ).capitalize().strip()

    print(f"Poor {user_name}, you have to get prepared to fight for your life.!")

    space()

## Attributes used to ensure a valide input and no endless game
    valid_input = False
    error_count = 0

## User has to enter the gender
    while(valid_input == False) :

## User has only two tries, otherwise the game is lost        
        if error_count >= 2:
            loose()
        else:
            pass
        
        gender = input(prompt = "What's your gender? M/F:  " ).strip().upper()

## Validate the Input, wheather it is Male / Female, to show who is the partner.
## And catch invalid inputs
        try:
            if gender in ['M', 'm', 'MALE']:
                print(f"Your partner from District 12 is Katniss Everdeen!")
                valid_input = True
            elif gender in ['F','f','FEMALE']:
                print(f"Your partner from District 12 is Peeta Mellark!")
                valid_input = True
## Catch if the user does not enter m, M, f nor F            
            else:
                raise ValueError("Invalid input")
                
## Catch improper inputs from an unexpected exit
        except ValueError as e:
            print(f"Sorry, I did not understand your input, please try again and check to only use M or F. You have only one more try left.")
        error_count = error_count + 1

    space()
    
## Show all opponents
    print_opponents()
    space()
    ## Logically seperate the stages with a small user input 
    input(prompt = "<Press Enter to Continue!>")

# Main function containing all three stages to win the Hunger Games.
def main():

## Attributes used to ensure a valide input and no endless game
    valid_input = False
    error_count = 0
    
    random_waiver = random.randint(1, 3)

## Introducrtion of the First Stage
    print(f"""
                      ________________  
                    |                  |
Welcome to Stage    |         1        |
                    | ________________ | 
                                """)
    
## While loop for multiple input tries after invalid inputs
    while(valid_input == False) :

## User can chosse the action for the first stage
        first_choice = input(prompt = f"""Your first stage is to get prepared. What would you like to do? Before going into the arena?
1. Go to the forest to expand your hunt?
2. Take the train to the capitol training center?
3. Go home and say goodbye to your Family?
Your choice: """).strip().upper()
#        print(f"Your input is: {first_choice}, the Rand_Waiver is: {random_waiver}")
        try:

## Block the check for the first choise, any hunting or training makes the player pass the stage
            if first_choice in ['1', 'ONE', 'FOREST', 'HUNT']:
                space()
                print("Great choice to improve your skills in the forest.")
## set valid, to exit the loop
                valid_input = True
                print("Great Job, you passed the first stage. The training in combination with your previous skills helped you pass!")
        
## Block the check for the first choise, any hunting or training makes the player pass the stage
            elif first_choice in ['2', 'TWO','TRAINING', 'CAPITOL', 'CENTER']:
                space()
                print("Great choice to go onto the train to the capitol. The training center is great to expand your skillset!")

## set valid, to exit the loop
                valid_input = True
                print("Great Job, you passed the first stage. The training in combination with your previous skills helped you pass!")

## Block the check for the first choise, no training makes the player loose the stage
            elif first_choice in ['3', 'THREE', 'HOME', 'FAMILY']:
                space()
## Check if you are randomly assigned a waiver
                if int(first_choice) == random_waiver:
                    print("Lucky you, you are assigned a random waiver to pass this stage!")
                else:
                    print("You're at home in your District. Your family is crying, but you stay strong.")
                    print("After a couple of minutes, you have to say goodbye and move on.")
## set valid, to exit the loop
#                    valid_input = True
                    print("Oh no,... you missed the training. You were expelled from the games. Try again")
                    loose()
                
                valid_input = True

## Invalid input increases the error count            
            else:
                space()
                print("Sorry, I could not understand your input, try again")
                                
## Catch improper inputs from an unexpected exit and increase error count      
        except ValueError:
            print("I could not read that, please try again.")
#            print(ValueError)

    ## The player has only two tries for a wrong input, afterwards I'll choose fot the player to pass by hunting
        if error_count >= 2:
            space()
            print("You have to choose something... I'll chose for you to go to say go hunting")
            print("Great choice to improve your skills in the forest.")
                
            ## set valid, to exit the loop
            valid_input = True
            print("Great Job, you passed the first stage. The training in combination with your previous skills helped you pass!")
        
        else:
            error_count = error_count + 1


## Logically seperate the stages with a small user input 
    space()
    input(prompt = "<Press Enter to Continue!>")
    
## Stage Two
    print(f"""
                      ________________  
                    |                  |
Welcome to Stage    |         2        |
                    | ________________ |
                    """)
    
    valid_input = False
    error_count = 0
    
    rand_hits = [random.randint(0, 9) for _ in range(4)]
    
## To master this stage the user has to choose a weapon and guess a number to pass.    
    while(valid_input == False) :
        try:
## User has only two tries, otherwise the game is lost        
            if error_count >= 2:
                loose()
            else:
                pass
## User inputs a weapn and a number
            print("You have to fight your first opponent. Choose a weapon and how often you want to hit.")
            second_weapon = input(prompt = "Choose between 'rock', 'box', 'axe', 'spear': ").strip().lower()
            second_number = int(input(prompt = "Enter a number between 1 and 10: ").strip())

## The following shows compinations how weapon and number to win / loos
            if second_weapon in ["rock", "r", "rok", "roc"]:
                if second_number <= rand_hits[0]:
                    if second_number < random.randint(2,5):
                        if second_number > 2:
                            print("You have survived the second stage!")
                            valid_input = True
                        else:
                            loose()
                    else:
                        loose()
                else:
                    print("You have survived the second stage!")
                    valid_input = True
        
## The following shows compinations how weapon and number to win / loos
            elif second_weapon in ["box", "b", "bx", "bo"]:
                if second_number <= rand_hits[1]:
                    loose()
                else:
                    print("You have survived the second stage!")
                    valid_input = True
        
## The following shows compinations how weapon and number to win / loos
            elif second_weapon in ["axe", "a", "ax", "ae"]:
                if second_number <= rand_hits[2]:
                    loose()
                else:
                    print("You have survived the second stage!")                
                    valid_input = True
        
## The following shows compinations how weapon and number to win / loos
            elif second_weapon in ["spear", "spr", "s", "spea"]:
                if second_number <= rand_hits[3]:
                    loose()
                else:
                    print("You have survived the second stage!")
                    valid_input = True

## Wrong input restarts the input
            else:
                space()
                print("I could not read that. Try again and choose between 'rock', 'box', 'axe', 'spear'.")

## Catch improper inputs from an unexpected exit and increase error count      
        except ValueError:
            print("I could not read that, please try again.")
        
        error_count = error_count + 1
## Logically seperate the stages with a small user input 
    space()
    input(prompt = "<Press Enter to Continue!>")
    
## Thrid Stage
    print(f"""
                      ________________  
                    |                  |
Welcome to Stage    |         3        |
                    | ________________ |
                    """)
    
## Attributes used to ensure a valide input and no endless game
    valid_input = False
    error_count = 0
    
    winning = random.randint(1, 2)
    
    while(valid_input == False) :
        ## User has only two tries, otherwise the game is lost        
        if error_count >= 2:
            loose()
        else:
            pass

## try and except to catch wrong inputs
        try:  
        
## user input for the final stage
            third_input = int(input(f"""You have come across a patch of deadly berries. What do you do?
1. Eat the berries and die.
2. Kill your partner from District 12.
Your choice: """).strip())
 
 ## Input 1 leads to win the game, 2 to loose the game (See Movie ;)
            if third_input in [1,2]:
                if third_input == winning:
                    win()
                    valid_input = True
                else:
                    loose()
                    valid_input = True
## Catch invalid input
            else:
                print("I could not read that, please try again.")
                
## Catch improper inputs from an unexpected exit            
        except ValueError:
            print("I could not read that, please try again.")
            
        error_count = error_count + 1


## Call the intro and main function to start            
intro()
main()
```
